```python
#  Python program for demonstrating protected members  
  
class Base1:  
    def __init__(self):  
  
        self._p = 78  
  
class Derived1(Base1):  
    def __init__(self):  
  
        Base1.__init__(self)  
        print ("We will call the protected member of base class: ",  
            self._p)  
  
        self._p = 433  
        print ("we will call the modified protected member outside the class: ",  
            self._p)  
  
  
obj_1 = Derived1()  
  
obj_2 = Base1()  
   
print ("Access the protected member of obj_1: ", obj_1._p)  
  
print ("Access the protected member of obj_2: ", obj_2._p)  
```

```python
# Python program for demonstrating private members

class Student:

   def __init__(self, name="Rajaram", marks=50):
      self.__name = name
      self.__marks = marks
   
   def name(self):
      return self.__name

   def marks(self):
      return self.__marks

   def studentdata(self):
      print("Name: {} marks: {}".format(self.__name, self.__marks))
      
s1 = Student()
s2 = Student("Bharat", 25)

s1.studentdata()
s2.studentdata()

# Accessing attributes using getter methods
print("Name: {} marks: {}".format(s1.name(), s1.marks()))
print("Name: {} marks: {}".format(s2.name(), s2.marks()))
```

```python
# Python program for public and private members demonstrating encapsulation

class Employee:
   
    def __init__(self, name, salary):
       
        self.name = name
        
        self.__salary = salary

emp = Employee('Jessa', 10000)

print('Salary:', emp.__salary)
```

```python
# public data members

class Employee:

    def __init__(self, name, salary):
        # public data members
        self.name = name
        self.salary = salary

    def show(self):
       
        print("Name: ", self.name, 'Salary:', self.salary)

emp = Employee('Jessa', 10000)

print("Name: ", emp.name, 'Salary:', emp.salary)

emp.show()
```

```python
# Proctecd member in inheritance

class Company:
    def __init__(self):
        # Protected member
        self._project = "NLP"

class Employee(Company):
    def __init__(self, name):
        self.name = name
        Company.__init__(self)

    def show(self):
        print("Employee name :", self.name)
       
        print("Working on project :", self._project)

c = Employee("Jessa")
c.show()

print('Project:', c._project)
```
